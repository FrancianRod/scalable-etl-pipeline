```python
import numpy as np
import pandas as pd
%run Caminhos.ipynb

tabela_base_gerada = pd.read_parquet(f'{pasta_hierarquia}{caminho_tabela_base_gerada_parquet}')
tabela_especificacao_ind_O_E = pd.read_parquet(f'{pasta_hierarquia}{caminho_especificacao_ind_O_E_parquet}')
```

BASE DE DADOS COMPARATIVA PAÍS

```python
# Filtro por escola
BASE_GERADA_pais = tabela_base_gerada[(tabela_base_gerada['Resultado Comparativo'].str.contains("SIM")) & (tabela_base_gerada['Hierarquia'] == "Pais")]

# Adicionar a nova coluna 'TipoIndicadorEstPro' com a informação 'Manual' em todas as linhas

BASE_GERADA_pais = BASE_GERADA_pais[['Codigo Indicador', 'Ano', 'Mes_Referente', 'Pais', 'Quantidade', 'Data Cadastro', 'Hierarquia']]

# Agrupar pelo 'Codigo Indicador', 'Ano', 'Mes_Referente' e pegar a linha com a data mais recente e o maior Mes_Referente
result_pais = BASE_GERADA_pais.loc[
    BASE_GERADA_pais.sort_values(by=['Data Cadastro', 'Mes_Referente'], ascending=[False, False])
    .groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])
    .head(1).index
]

result_pais['Codigo Indicador'] = result_pais['Codigo Indicador'].astype(int)

# Renomear duas colunas ao mesmo tempo
result_pais = result_pais.rename(columns={'Quantidade': 'Quantidade País'})

# Agrupar pelo 'Codigo Indicador', 'Ano', e 'Mes_Referente', e pegar o índice das linhas com a data mais recente
result_pais = result_pais.loc[result_pais.groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])['Data Cadastro'].idxmax()]

result_pais = result_pais.drop(['Mes_Referente', 'Data Cadastro', 'Hierarquia'], axis=1)

# Exibir o DataFrame
display(result_pais)
```

BASE DE DADOS COMPARATIVA REGIÃO

```python
# Filtro por escola
BASE_GERADA_regiao = tabela_base_gerada[(tabela_base_gerada['Resultado Comparativo'].str.contains("SIM")) & (tabela_base_gerada['Hierarquia'] == "Regiao")]

# Adicionar a nova coluna 'TipoIndicadorEstPro' com a informação 'Manual' em todas as linhas

BASE_GERADA_regiao = BASE_GERADA_regiao[['Codigo Indicador', 'Ano', 'Mes_Referente', 'Regiao', 'Quantidade', 'Data Cadastro', 'Hierarquia']]

# Agrupar pelo 'Codigo Indicador', 'Ano', 'Mes_Referente' e pegar a linha com a data mais recente e o maior Mes_Referente
result_regiao = BASE_GERADA_regiao.loc[
    BASE_GERADA_regiao.sort_values(by=['Data Cadastro', 'Mes_Referente'], ascending=[False, False])
    .groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])
    .head(1).index
]

result_regiao['Codigo Indicador'] = result_regiao['Codigo Indicador'].astype(int)

# Renomear duas colunas ao mesmo tempo
result_regiao = result_regiao.rename(columns={'Quantidade': 'Quantidade Regiao'})

# Agrupar pelo 'Codigo Indicador', 'Ano', e 'Mes_Referente', e pegar o índice das linhas com a data mais recente
result_regiao = result_regiao.loc[result_regiao.groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])['Data Cadastro'].idxmax()]

result_regiao = result_regiao.drop(['Mes_Referente', 'Data Cadastro', 'Hierarquia'], axis=1)

# Exibir o DataFrame
display(result_regiao)
```

BASE DE DADOS COMPARATIVA - MELHOR ESTADO DO BRASIL

```python
# Filtro por escola
BASE_GERADA = tabela_base_gerada[(tabela_base_gerada['Resultado Comparativo'].str.contains("SIM")) & (tabela_base_gerada['Hierarquia'] == "Estado")]

# Adicionar a nova coluna 'TipoIndicadorEstPro' com a informação 'Manual' em todas as linhas

BASE_GERADA = BASE_GERADA[['Codigo Indicador', 'Ano', 'Mes_Referente', 'Id Estado', 'Estado', 'Quantidade', 'Data Cadastro', 'Hierarquia']]

# Agrupar pelo 'Codigo Indicador', 'Ano', 'Mes_Referente' e pegar a linha com a data mais recente e o maior Mes_Referente
melhor_result_estado = BASE_GERADA.loc[
    BASE_GERADA.sort_values(by=['Data Cadastro', 'Mes_Referente'], ascending=[False, False])
    .groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])
    .head(1).index
]

melhor_result_estado['Codigo Indicador'] = melhor_result_estado['Codigo Indicador'].astype(int)

# Renomear duas colunas ao mesmo tempo
melhor_result_estado = melhor_result_estado.rename(columns={'Estado': 'Melhor Estado', 'Quantidade': 'Melhor Quantidade'})

# Agrupar pelo 'Codigo Indicador', 'Ano', e 'Mes_Referente', e pegar o índice das linhas com a data mais recente
melhor_result_estado = melhor_result_estado.loc[melhor_result_estado.groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])['Data Cadastro'].idxmax()]

melhor_result_estado = melhor_result_estado.drop(['Mes_Referente', 'Data Cadastro', 'Hierarquia'], axis=1)
#melhor_result_estado = melhor_result_estado[(melhor_result_estado["Codigo Indicador"] == 2266)]
# Exibir o DataFrame
display(melhor_result_estado)

# salvar os dados em um arquivo csv
#%timeit melhor_result_estado.to_csv(f'{pasta_hierarquia}{caminho_teste1}')
```

BASE DE DADOS COMPARATIVA ESTADO

```python
# Filtro por escola
BASE_GERADA = tabela_base_gerada[(tabela_base_gerada['Resultado Comparativo'].str.contains("NÃO")) & (tabela_base_gerada['Hierarquia'] == "Estado")]

# Adicionar a nova coluna 'TipoIndicadorEstPro' com a informação 'Manual' em todas as linhas

BASE_GERADA = BASE_GERADA[['Codigo Indicador', 'Ano', 'Mes_Referente', 'Id Estado', 'Estado', 'Quantidade', 'Data Cadastro', 'Hierarquia']]

# Agrupar pelo 'Codigo Indicador', 'Ano', 'Mes_Referente' e pegar a linha com a data mais recente e o maior Mes_Referente
result_piaui = BASE_GERADA.loc[
    BASE_GERADA.sort_values(by=['Data Cadastro', 'Mes_Referente'], ascending=[False, False])
    .groupby(['Codigo Indicador', 'Ano', 'Hierarquia'])
    .head(1).index
]

result_piaui['Codigo Indicador'] = result_piaui['Codigo Indicador'].astype(int)

result_piaui = result_piaui.drop(['Mes_Referente', 'Data Cadastro', 'Hierarquia'], axis=1)
#result_piaui = result_piaui[(result_piaui["Codigo Indicador"] == 2266)]

# Exibir o DataFrame
display(result_piaui)

# salvar os dados em um arquivo csv
#%timeit melhor_result_estado.to_csv(f'{pasta_hierarquia}{caminho_teste2}')
```

Melhor estado atualizado

```python
# Filtrando os códigos que estão em `melhor_result_estado`
codigos_melhor_estado = melhor_result_estado['Codigo Indicador'].unique()

# Excluindo esses códigos de `result_piaui`
result_piaui_filtrado = result_piaui[~result_piaui['Codigo Indicador'].isin(codigos_melhor_estado)]

# Concatenando as duas tabelas
final_df = pd.concat([melhor_result_estado, result_piaui_filtrado], ignore_index=True)

# Removendo duplicatas, priorizando o "Resultado Melhor Estado"
final_df = final_df.drop_duplicates(keep='first')

# Preenchendo valores ausentes em "Melhor Estado" com os valores da coluna "Estado"
final_df['Melhor Estado'] = final_df['Melhor Estado'].fillna(final_df['Estado'])
final_df['Melhor Quantidade'] = final_df['Melhor Quantidade'].fillna(final_df['Quantidade'])

# Removendo a coluna "Estado", já que a informação agora está consolidada em "Melhor Estado"
final_df = final_df.drop(columns=['Estado', 'Quantidade'])
#final_df = final_df[(final_df["Codigo Indicador"] == 2266)]
display(final_df)
# salvar os dados em um arquivo csv
#%timeit final_df.to_csv(f'{pasta_hierarquia}{caminho_teste2}')
```

Tabela Comparativa

```python
# Supondo que temos dois DataFrames: df1 e df2

tab_comparativa = result_piaui.merge(final_df, on=['Codigo Indicador', 'Ano'], how='left')
tab_comparativa = tab_comparativa.merge(result_regiao, on=['Codigo Indicador', 'Ano'], how='left')
tab_comparativa = tab_comparativa.merge(result_pais, on=['Codigo Indicador', 'Ano'], how='left')

#Avaliar ERRO relacionado a coluna 'Id Estado'

display(tab_comparativa)
```

Tabela de Especificação de Indicadores

```python
tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E[["Codigo Indicador", "MetaMes", "TipoIndicadorEstPro", "UnidadeMedida", "Ano_Meta", "Mes_Meta", "GlossarioIndicador"]]

# Agrupar pelo 'Codigo Indicador', 'Ano', 'Mes_Referente' e pegar a linha com a data mais recente e o maior Mes_Referente
tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E.loc[
    tabela_especificacao_ind_O_E.sort_values(by=['Mes_Meta'], ascending=[False])
    .groupby(['Codigo Indicador', 'Ano_Meta'])
    .head(1).index
]

# Verificando as colunas antes de renomear
#tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E[tabela_especificacao_ind_O_E['Codigo Indicador'] == 177]
#display(tabela_especificacao_ind_O_E)
# Renomear a coluna de contagem para diferenciar
tabela_especificacao_ind_O_E.rename(columns={'Ano_Meta': 'Ano'}, inplace=True)

tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E[tabela_especificacao_ind_O_E['GlossarioIndicador'].str.contains("Compromisso", case=False)]

# Filtra o DataFrame removendo as linhas que contenham "EXCLUIR" na coluna GlossarioIndicador
tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E[~tabela_especificacao_ind_O_E['GlossarioIndicador'].str.contains("EXCLUIR", na=False)]

tabela_especificacao_ind_O_E = tabela_especificacao_ind_O_E.drop_duplicates()

# Mesclar os DataFrames para adicionar as colunas com o cálculo
tab_comparativo_gov = tabela_especificacao_ind_O_E.merge(tab_comparativa, on=['Codigo Indicador', 'Ano'], how='left')

tab_comparativo_gov['MetaMes'] = tab_comparativo_gov['MetaMes'].fillna(np.nan)
#tab_comparativo_gov = tab_comparativo_gov[(tab_comparativo_gov["Codigo Indicador"] == 73)]
#tab_comparativo_gov = tab_comparativo_gov[(tab_comparativo_gov["Ano_Meta"] == 2022)]

display(tab_comparativo_gov)

tab_comparativo_gov.to_csv(f'{pasta_hierarquia}{caminho_base_comparativa_csv}', index=False)
tab_comparativo_gov.to_parquet(f'{pasta_hierarquia}{caminho_base_comparativa_parquet}', index=False)
```
