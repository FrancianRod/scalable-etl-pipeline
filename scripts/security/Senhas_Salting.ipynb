Análise Hashing - Verificação Senha

```python
import bcrypt

# Criar hash da senha
senha = b"minhaSenha123"  # bcrypt exige bytes
salt = bcrypt.gensalt()  # Gera um sal seguro automaticamente
hash_senha = bcrypt.hashpw(senha, salt)

print("Hash bcrypt:", hash_senha)

# Verificar senha digitada
senha_digitada = b"minhaSenha123"
if bcrypt.checkpw(senha_digitada, hash_senha):
    print("✅ Senha correta!")
else:
    print("❌ Senha incorreta!")
```

```python
import bcrypt

# Criar hash da senha (simulando um cadastro)
senha = b"minhaSenha123"  # bcrypt exige bytes
salt = bcrypt.gensalt()  # Gera um sal seguro automaticamente
hash_senha = bcrypt.hashpw(senha, salt)

print("\n🔒 Senha cadastrada com sucesso!")
print("Hash bcrypt:", hash_senha)

# Simular um login pedindo a senha ao usuário
senha_digitada = input("\n🔑 Digite sua senha para login: ").encode()  # Convertendo para bytes

# Verificar se a senha digitada está correta
if bcrypt.checkpw(senha_digitada, hash_senha):
    print("\n✅ Login bem-sucedido! A senha está correta.")
else:
    print("\n❌ Login falhou! Senha incorreta.")
```

Cadastro e Verificação de Login e Senha com Salting e bcrypt

```python
import bcrypt
import sqlite3

# Conectar ao banco (ou criar se não existir)
conn = sqlite3.connect("usuarios.db")
cursor = conn.cursor()

# Criar tabela para armazenar usuários (se não existir)
cursor.execute("""
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    username TEXT UNIQUE NOT NULL,
    senha_hash BLOB NOT NULL
)
""")
conn.commit()

# ----------------- CADASTRO DO USUÁRIO -----------------
usuario = input("\n👤 Digite seu nome de usuário: ")
senha = input("🔑 Digite sua senha: ").encode()  # Convertendo para bytes

# Gerar hash da senha
salt = bcrypt.gensalt()
hash_senha = bcrypt.hashpw(senha, salt)

# Armazenar no banco de dados
cursor.execute("INSERT INTO usuarios (username, senha_hash) VALUES (?, ?)", (usuario, hash_senha))
conn.commit()
print("\n✅ Usuário cadastrado com sucesso!\n")

# ----------------- LOGIN -----------------
usuario_login = input("👤 Usuário: ")
senha_digitada = input("🔑 Digite sua senha para login: ").encode()

# Recuperar o hash armazenado no banco
cursor.execute("SELECT senha_hash FROM usuarios WHERE username = ?", (usuario_login,))
resultado = cursor.fetchone()

if resultado:
    hash_armazenado = resultado[0]

    # Verificar se a senha digitada gera o mesmo hash
    if bcrypt.checkpw(senha_digitada, hash_armazenado):
        print("\n✅ Login bem-sucedido! A senha está correta.")
    else:
        print("\n❌ Login falhou! Senha incorreta.")
else:
    print("\n❌ Usuário não encontrado.")

# Fechar conexão com o banco
conn.close()
```

Verificar o Banco Salvo

```python
import sqlite3

# Conectar ao banco de dados
conn = sqlite3.connect("usuarios.db")
cursor = conn.cursor()

# Recuperar todos os usuários cadastrados
cursor.execute("SELECT id, username, senha_hash FROM usuarios")
usuarios = cursor.fetchall()

# Exibir os usuários cadastrados
print("\n👥 Usuários cadastrados:")
for usuario in usuarios:
    print(f"🆔 ID: {usuario[0]} | 👤 Usuário: {usuario[1]} | 🔐 Hash: {usuario[2]}")

# Fechar conexão com o banco
conn.close()
```
